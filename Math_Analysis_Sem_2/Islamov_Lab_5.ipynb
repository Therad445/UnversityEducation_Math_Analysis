```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
def m(x0,y0,ak,bk,n0,e):
    n1=2*n0
    n=0
    t=True
    a=[]
    b=[]
    c=[]
    d=[]
    while True:
        a=[x0]
        b=[y0]
        c=[x0]
        d=[y0]
        dx1=(bk-ak)/(n0)
        dx2=(bk-ak)/(n1)
        maxl=0
        for i in range(1,n0):
            a.append(a[i-1]+dx1)
            b.append(b[i-1]+f(a[i-1],b[i-1])*dx1)
        for i in range(1,n1):
            c.append(c[i-1]+dx2)
            d.append(d[i-1]+f(c[i-1],d[i-1])*dx2)
        for i in range(1,n1):
            for j in range(1,n0):
                if c[i]==a[j]:
                    razn=abs(d[i]-b[j])
                    if razn>maxl:
                        maxl=razn
        if t:
            for i in range(n0-1):
                plt.plot([a[i],a[i+1]],[b[i],b[i+1]],"b-")
            for i in range(n1-1):
                plt.plot([c[i],c[i+1]],[d[i],d[i+1]],"r-")
            t=False
        if maxl<e:
            n=n1 
            for i in range(n-1):
                plt.plot([c[i],c[i+1]],[d[i],d[i+1]],"g-")
            break 
        else:
            n0=n1
            n1=2*n1
    plt.grid()
    print(n)
    return [c,d]
def f(x,y):
    return x*y
mas=m(0,1,0,1,4,0.001)
```

<big><big><b> 1) Метод ломаных Эйлера (или метод Эйлера с постоянным шагом) является численным методом для приближенного решения обыкновенных дифференциальных уравнений (ОДУ) первого порядка. Суть метода заключается в аппроксимации кривой решения ОДУ с помощью ломаных линий. Шаг метода фиксирован и равен h.

<big><big><b> Шаги метода ломаных Эйлера:
- Задается начальное условие y(x₀) = y₀, где y₀ - известное значение функции y в начальной точке x₀.
- Выбирается фиксированный шаг h.
- Последовательно вычисляются значения функции y(x) в узлах сетки xₖ = x₀ + kh, где k - номер шага.
- На каждом шаге используется линейная аппроксимация касательной к кривой решения ОДУ в текущей точке.
- Для вычисления значения функции y(xₖ₊₁) на следующем шаге применяется формула: y(xₖ₊₁) = y(xₖ) + h * f(xₖ, y(xₖ)), где f(x, y) - правая часть ОДУ.

<big><big><b> Метод ломаных Эйлера является простым, но обычно имеет второй порядок точности и может быть нестабилен для некоторых ОДУ с большими значениями производных.

<big><big><b> 2) Метод последовательных приближений (итераций) является численным методом для решения уравнений, в том числе нелинейных. Суть метода заключается в последовательном приближении к решению путем итеративного применения функционального оператора к начальному приближению.

<big><big><b> Шаги метода последовательных приближений:
- Задается начальное приближение x₀.
- Последовательно применяется функциональный оператор g к текущему приближению: xₖ₊₁ = g(xₖ), где k - номер итерации.
- Итеративный процесс продолжается до достижения требуемой точности или установления сходимости.
- Условие сходимости может быть задано, например, как ||xₖ₊₁ - xₖ|| < ε, где ε - заданная точность и ||.|| - некоторая норма.

<big><big><b> Выбор функционального оператора g зависит от конкретного уравнения и требует анализа свойств решения. Метод последовательных приближений особенно полезен для нелинейных уравнений, где отсутствуют аналитические методы решения.
