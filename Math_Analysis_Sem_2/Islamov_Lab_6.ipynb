```python
#а) Верхняя и нижняя границы сплошные:

import numpy as np
import matplotlib.pyplot as plt

def plot_functions_A(f, g, a, b, n):
    x = np.linspace(a, b, n)
    y_f = f(x)
    y_g = g(x)
    
    plt.fill_between(x, y_f, y_g, where=(y_f>=y_g), color='gray')
    plt.plot(x, y_f, label='y = f(x)')
    plt.plot(x, y_g, label='y = g(x)')
    plt.grid()
    plt.axhline(y=0, color="k")
    plt.axvline(x=0, color="k")
    plt.legend()
    plt.show()


#б) Обе границы пунктирные:

def plot_functions_B(f, g, a, b, n):
    x = np.linspace(a, b, n)
    y_f = f(x)
    y_g = g(x)
    
    plt.fill_between(x, y_f, y_g, where=(y_f>=y_g), color='gray')
    plt.plot(x, y_f, label='y = f(x)', linestyle='--')
    plt.plot(x, y_g, label='y = g(x)', linestyle='--')
    plt.grid()
    plt.axhline(y=0, color="k")
    plt.axvline(x=0, color="k")
    plt.legend()
    plt.show()


#в) Одна граница сплошная, другая пунктирная:

def plot_functions_V(f, g, a, b, n):
    x = np.linspace(a, b, n)
    y_f = f(x)
    y_g = g(x)
    
    plt.fill_between(x, y_f, y_g, where=(y_f>=y_g), color='gray')
    plt.plot(x, y_f, label='y = f(x)', linestyle='-')
    plt.plot(x, y_g, label='y = g(x)', linestyle='--')
    plt.grid()
    plt.axhline(y=0, color="k")
    plt.axvline(x=0, color="k")
    plt.legend()
    plt.show()
    
    
def f(x):
    return np.sin(x)

def g(x):
    return np.cos(x)

plot_functions_V(f, g, 0, 4, 100)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x, y):
    return 1/x+1/y
def plot_3d(f, a, b):
    x_min, x_max = a, b
    y_min, y_max = a, b

    X, Y = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))

    # Вычисление значений функции на сетке точек
    Z = f(X, Y)

    fig = plt.figure()
    ax = plt.axes(projection='3d')
    wf = ax.plot_wireframe(X, Y, Z, rstride = 10, cstride = 10, color = 'green')
    wf.set_edgecolors(wf.to_rgba(wf._A))
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    fig.colorbar(wf, shrink=0.5, aspect=5)
    ax.view_init(15, 70)
    plt.show()
plot_3d(f, 0.1, 2)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return 1/x+1/y

a = 0.1
b = 2
n = 50

x = np.linspace(a, b, n)
y = np.linspace(a, b, n)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# построение каркасной поверхности
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

# линии уровня без значений
cset = ax.contour(X, Y, Z, zdir='z', offset=-0.5)
cset = ax.contour(X, Y, Z, zdir='x', offset=a)
cset = ax.contour(X, Y, Z, zdir='y', offset=b)

# линии уровня с значениями
cset = ax.contour(X, Y, Z, zdir='z', offset=-0.5, linewidths=2, cmap='viridis')
cset = ax.contour(X, Y, Z, zdir='x', offset=a, linewidths=2, cmap='viridis')
cset = ax.contour(X, Y, Z, zdir='y', offset=b, linewidths=2, cmap='viridis')

# добавление значений на линии уровня с шагом 0.1
ax.clabel(cset, fontsize=10, inline=True, fmt='%.1f')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.grid()
plt.show()
```
