```python
def rd_metrdim(x,y):
    n = len(x)
    mas = list(range(n))
    for k in range(n):
        mas[k] = abs(x[k]-y[k])
    R0 = max(mas)
    R1 = sum(mas)
    R2 = 0
    for k in range(n):
        R2 += (abs(x[k]-y[k]))**2
    R2 = (R2)**(1/2)
    return R0, R1, int(R2)
    
print(rd_metrdim([0, 0], [4, 3]))
print(rd_metrdim([1,2,3,4], [7, 3, 4, 200]))
```

```python

import numpy as np
import matplotlib.pyplot as plt

def rho0(x, y):
    """Метрика ρ0"""
    return max(abs(x), abs(y))

def rho1(x, y):
    """Метрика ρ1"""
    return abs(x) + abs(y)

def rho2(x, y):
    """Метрика ρ2"""
    return np.sqrt(x**2 + y**2)

def plot_ball(metric,r,x0):
    """Построение шара BO,1 в заданной метрике"""
    n = len(x0)
    x = np.linspace(-r+x0[0], r+x0[0], 500)
    y = np.linspace(-r+x0[1], r+x0[1], 500)
    X, Y = np.meshgrid(x, y)
    Z = np.zeros_like(X)
    for i in range(len(x)):
        for j in range(len(y)):
            if metric(X[i,j]-x0[0], Y[i,j]-x0[1]) <= r:
                Z[i,j] = r
    plt.contourf(X, Y, Z, cmap='Reds')
    plt.axis('equal')
    plt.axhline(y=0, color="k")
    plt.axvline(x=0, color="k")
    plt.show()

plot_ball(rho0,1,[0,0])

plot_ball(rho1,1,[0,0])

plot_ball(rho2,1,[0,0])
```

```python
import numpy as np

import numpy as np

def iter_method(A, B, eps):
    n = len(A)
    X0 = B.copy()
    X1 = np.zeros(n)
    k = 0
    while np.linalg.norm(X1 - X0) > eps:
        X0 = X1.copy()
        for i in range(n):
            s = 0
            for j in range(n):
                if i != j:
                    s += A[i][j] * X1[j]
            X1[i] = (B[i] - s) / A[i][i]
        k += 1
    return X1, k

A = np.array([[-0.23, -9.31, 1.41], [-1.89, -1.24, -10.27], [-8.41, -0.23, -1.67]])
B = np.array([[4.26], [-7.51], [8.23]])

"""
A = [[1, 0.06, -0.02],
     [0.03, 1, -0.05],
     [0.01, -0.02, 1]]

B = [2, 3, 5]

"""
e = 0.001
x, iters = iter_method(A, B, e)

print("Приближенное решение:", x)
print("Количество итераций:", iters)
print("Матрица С не сжимающая")
```

<big><big><b> 1. Принцип сжимающих отображений утверждает, что если отображение f: X -> X (где X - непустое множество) является сжимающим на множестве X, то оно имеет единственную неподвижную точку. Здесь сжимающее отображение означает, что для любых двух точек x и y из множества X выполнено неравенство d(f(x), f(y)) <= q * d(x, y), где d обозначает метрику на множестве X, а q - константа с 0 <= q < 1. Этот принцип широко применяется при решении уравнений и в численных методах.
