```python
import numpy as np
from scipy import integrate
from scipy.optimize import fsolve
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x * np.cos(x)

x = np.linspace(0, np.pi/2, 100)
y = f(x)

plt.plot(x, y)
plt.fill_between(x, y, where=(y>0), color='gray', alpha=0.5)
plt.grid()
plt.xlim(0, np.pi/2)
plt.ylim(0, 1.5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('$y=x\cos(x)$')
plt.show()

S = integrate.simpson(x * np.cos(x), x)
print(f'Площадь фигуры: {S:.4f}')
```

```python
def f1(x):
    return x*x + 2*x

def f2(x):
    return 7 - 4*x - x*x


def equations(p):
    x, y = p
    return (x*x + 2*x - y, 7 - 4*x - x*x - y)

x = np.linspace(-5, 3, 100)
y1 = f1(x)
y2 = f2(x)

plt.plot(x, y1, label='y=x^2+2x')
plt.plot(x, y2, label='y=7-4x-x^2')
plt.fill_between(x, y1, y2, where = (y2 > y1), color='gray', alpha = 0.5)
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Графики функций')
plt.grid()
plt.show()



x1, y1 = fsolve(equations, (-4, 16))
x2, y2 = fsolve(equations, (0, 7))

print('Точки пересечения графиков:')
print(f'({x1:.2f}, {y1:.2f})')
print(f'({x2:.2f}, {y2:.2f})')

x = np.linspace(x1, x2, 100)
S = integrate.simpson(f2(x) - f1(x), x)

print(f"Площадь: {S:.4f}")
```

```python
import numpy as np
from scipy.integrate import quad

# Уравнение кардиоиды
def rho(phi):
    return 3 * (1 - np.cos(phi))

# Интегрируемая функция для вычисления дифференциала дуги
def integrand(phi):
    return np.sqrt(rho(phi)**2 + rho(phi + 0.01)**2)

# Вычисляем длину дуги
arc_length, _ = quad(integrand, 0, 2*np.pi)

print(f"Длина дуги кардиоиды: {arc_length}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Создаем массивы для наших функций
x = np.linspace(-10, 10, 1000)
y1 = x**2 + 1
y2 = 3 - x
y3 = np.ones_like(x)

# Создаем график и добавляем наши функции
fig, ax = plt.subplots()
ax.plot(x, y1, label='y=x^2+1')
ax.plot(x, y2, label='y=3-x')
ax.plot(x, y3, label='y=1')

# Вычисляем точки пересечения графиков
x_intersect1 = (3 + np.sqrt(5))/2
x_intersect2 = (3 - np.sqrt(5))/2

# Вычисляем площадь фигуры
x_fill = np.linspace(x_intersect2, x_intersect1, 1000)
y_fill1 = x_fill**2 + 1
y_fill2 = 3 - x_fill
ax.fill_between(x_fill, y_fill1, y_fill2, where=y_fill1>y_fill2, color='gray', alpha=0.5)

# Добавляем сетку и метки осей
ax.grid()
ax.set_xlabel('x')
ax.set_ylabel('y')

# Добавляем заголовок и легенду
ax.set_title('Графики функций')
ax.legend()

# Устанавливаем одинаковый масштаб по осям
ax.set_aspect('equal')

# Отображаем график
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Задание углов и радиусов
theta = np.linspace(0, 2*np.pi, 1000)
r1 = 1
r2 = 2
sigma1 = np.pi/4
sigma2 = 3*np.pi/4

# Вычисление координат точек
x1 = r1 * np.cos(theta)
y1 = r1 * np.sin(theta)

x2 = r2 * np.cos(theta)
y2 = r2 * np.sin(theta)

x3 = np.cos(sigma1) * r1
y3 = np.sin(sigma1) * r1

x4 = np.cos(sigma2) * r1
y4 = np.sin(sigma2) * r1

# Построение фигуры
plt.plot(x1, y1, label='r = 1')
plt.plot(x2, y2, label='r = 2')
plt.plot([0, x3], [0, y3], label='sigma = pi/4')
plt.plot([0, x4], [0, y4], label='sigma = 3pi/4')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Фигура, ограниченная кривыми и лучами')
plt.legend()
plt.grid(True)

# Вычисление площади фигуры
area_sector = (sigma2 - sigma1) * (r1**2) / 2
area_circle1 = np.pi * (r1**2)
area_circle2 = np.pi * (r2**2)
area = area_circle2 - area_circle1 - area_sector

# Вывод площади фигуры
print(f"Площадь фигуры: {area}")

plt.show()
```

```python
import numpy as np

# Задание параметрических уравнений
def x(t):
    return np.cos(t)**3

def y(t):
    return np.sin(t)**3

# Задание пределов интегрирования
a = 0
b = 2 * np.pi

# Функция для вычисления длины дуги
def arc_length(x, y, a, b):
    def integrand(t):
        dx_dt = 3 * np.cos(t)**2 * (-np.sin(t))
        dy_dt = 3 * np.sin(t)**2 * np.cos(t)
        return np.sqrt(dx_dt**2 + dy_dt**2)

    t_values = np.linspace(a, b, 1000)
    dt = (b - a) / 1000
    arc_length = np.sum(integrand(t_values)) * dt

    return arc_length

# Вычисление длины дуги
length = arc_length(x, y, a, b)

# Вывод длины дуги
print(f"Длина замкнутой кривой: {length}")
```
