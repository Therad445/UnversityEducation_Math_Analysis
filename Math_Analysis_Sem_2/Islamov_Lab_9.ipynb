```python
import numpy as np
import matplotlib.pyplot as plt

# Определение функции
def f(x, y):
    return (1/3)*x**3 + (1/3)*y**3

# Определение ограничения
def constraint(x, y):
    return 1 - x - 4*y

# Создание сетки значений x и y
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)

# Вычисление значений z и ограничения
Z = f(X, Y)
C = constraint(X, Y)

# Построение поверхности
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, alpha=0.8, cmap='viridis')

# Построение кривой ограничения
ax.contour(X, Y, C, [0], colors='red')

# Поиск точек экстремума
extremum_points = np.argwhere(C == 0)
extremum_x = extremum_points[:, 1]
extremum_y = extremum_points[:, 0]
extremum_z = f(extremum_x, extremum_y)

# Отображение точек экстремума
ax.scatter(extremum_x, extremum_y, extremum_z, color='black', s=50, label='Условный экстремум')

# Установка меток осей и заголовка
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.title("z = (1/3)*x^3 + (1/3)*y^3, если 1 = x + 4y")

# Отображение графика
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Определение функции
def f(x, y):
    return (x+1)**2 + y**2

# Определение ограничения
def constraint(x, y):
    return y**2 - x**3

# Создание сетки значений x и y
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)

# Вычисление значений z и ограничения
Z = f(X, Y)
C = constraint(X, Y)

# Построение поверхности
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, alpha=0.8, cmap='viridis')

# Построение кривой ограничения
ax.contour(X, Y, C, [0], colors='red')

# Поиск точек экстремума
extremum_points = np.argwhere(C == 0)
extremum_x = extremum_points[:, 1]
extremum_y = extremum_points[:, 0]
extremum_z = f(extremum_x, extremum_y)

# Отображение точек экстремума
ax.scatter(extremum_x, extremum_y, extremum_z, color='black', s=50, label='Условный экстремум')

# Установка меток осей и заголовка
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.title("z = (x+1)^2 + y^2, если y^2 - x^3 = 0")

# Отображение графика
plt.legend()
plt.show()
```

```python
from sympy import symbols, solve

# Определение символьных переменных
x, y, λ = symbols('x y λ')

# Определение функции L(x, y, λ)
L = (1/3)*x**3 + (1/3)*y**3 - λ*(1 - x - 4*y)

# Нахождение частных производных и приравнивание к нулю
partial_x = L.diff(x)
partial_y = L.diff(y)
partial_λ = L.diff(λ)

# Решение системы уравнений
solutions = solve((partial_x, partial_y, partial_λ), (x, y, λ))

# Вывод решений
for solution in solutions:
    x_val = solution[0]
    y_val = solution[1]
    z_val = (1/3)*x_val**3 + (1/3)*y_val**3
    print(f"Точка условного экстремума: (x={x_val}, y={y_val}, z={z_val})")
```

```python
import sympy as sp

# Определение символьных переменных
x, y, l = sp.symbols('x y l')

# Проверка задачи a)
z_a = (1/3)*x**3 + (1/3)*y**3
constraint_a = x + 4*y - 1

# Проверка возможности применения метода множителей Лагранжа для задачи a)
lagrange_possible_a = sp.solve([sp.Eq(sp.diff(z_a, var), l*sp.diff(constraint_a, var)) for var in (x, y)], (x, y, l))

# Проверка задачи б)
z_b = (x+1)**2 + y**2
constraint_b = y**2 - x**3

# Проверка возможности применения метода множителей Лагранжа для задачи б)
lagrange_possible_b = sp.solve([sp.Eq(sp.diff(z_b, var), l*sp.diff(constraint_b, var)) for var in (x, y)], (x, y, l))

# Вывод результатов
if lagrange_possible_a:
    print("Метод множителей Лагранжа применим для задачи a)")
else:
    print("Метод множителей Лагранжа не применим для задачи a)")

if lagrange_possible_b:
    print("Метод множителей Лагранжа применим для задачи б)")
else:
    print("Метод множителей Лагранжа не применим для задачи б)")
```

```python
from sympy import symbols, solve

# Определение символьных переменных
x, y, l = symbols('x y l')

# Определение функции и ограничения
f = 1/3 * x**3 + 1/3 * y**3
phi = 1 - x - 4 * y

# Вычисление лагранжиана
L = f + l * phi

# Вычисление частных производных
dx = L.diff(x)
dy = L.diff(y)
dl = L.diff(l)

# Решение системы уравнений
solutions = solve([dx, dy, dl, phi], [x, y, l], dict=True)

# Вывод результатов
print("Найдено критических точек:", len(solutions))
for s in solutions:
    print(f"x={s[x]}, y={s[y]}")
    H = L.diff(x, x) + 2 * L.diff(x, y) + L.diff(y, y)
    H_ = H.subs({x: s[x], y: s[y], l: s[l]})
    if H_ > 0:
        print("Условный минимум:", s)
    elif H_ < 0:
        print("Условный максимум:", s)
    elif H_ == 0:
        print("Необходимо провести более детальный анализ:", s)
    else:
        print("Точка не является экстремумом:", s)
```

<big><big><b> 1) Прямой метод отыскания точек условного экстремума является одним из методов нахождения условных экстремумов функции с ограничениями. Суть метода заключается в поиске точек, где градиент функции и градиент ограничений находятся в линейной зависимости. Эти точки могут быть потенциальными кандидатами на условные экстремумы.

<big><big><b> 2) Необходимое условие условного экстремума функции двух переменных гласит: если функция f(x, y) достигает условного экстремума в точке (a, b) при ограничении g(x, y) = 0, и в этой точке существуют частные производные f и g, то градиент функции f(x, y) и градиент ограничения g(x, y) должны быть коллинеарными в этой точке. Формально это выражается следующим образом: ∇f(a, b) = λ∇g(a, b), где λ - множитель Лагранжа, который определяется из условия g(a, b) = 0.

<big><big><b> 3) Достаточное условие условного экстремума функции двух переменных гласит: если функции f(x, y) и g(x, y) имеют непрерывные частные производные в окрестности точки (a, b), и (a, b) является стационарной точкой лагранжиана L(x, y, λ) = f(x, y) - λg(x, y) при условии g(a, b) = 0, то следующая матрица, называемая матрицей Гессеана лагранжиана, должна иметь определитель отличный от нуля: 

| ∂²L/∂x²(a, b, λ)  ∂²L/(∂x∂y)(a, b, λ) |
| ∂²L/(∂y∂x)(a, b, λ)  ∂²L/∂y²(a, b, λ) |

<big><big><b> Если матрица Гессеана имеет положительный определитель, то точка (a, b) является локальным минимумом лагранжиана, а значит, потенциальным условным экстремумом функции f(x, y) при условии g(x, y) = 0. Если матрица Гессеана имеет отрицательный определитель, то точка (a, b) является локальным максимумом лагранжиана.
