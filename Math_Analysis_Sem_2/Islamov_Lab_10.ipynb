<big><big><big><b> <center> ## Краткие теоретические сведения и практические упражнения

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import dblquad

# Определение функции подынтегрального выражения
def f(x, y):
    return (x + y)**2

# Определение пределов интегрирования
x_lower = 0
x_upper = 2

# Функции, определяющие верхнюю и нижнюю границы интегрирования по y
def y_lower(x):
    return x**2

def y_upper(x):
    return 2 * x

# Визуализация области интегрирования
x = np.linspace(x_lower, x_upper, 100)
y1 = y_lower(x)
y2 = y_upper(x)

plt.plot(x, y1, 'r', label='y = x^2')
plt.plot(x, y2, 'b', label='y = 2x')
plt.fill_between(x, y1, y2, alpha=0.2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Область интегрирования')
plt.legend()
plt.grid(True)
plt.show()

# Вычисление двойного интеграла
integral, error = dblquad(f, x_lower, x_upper, y_lower, y_upper)

# Вывод результата
print("Значение двойного интеграла:", integral)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.integrate import tplquad

# Определение функции подынтегрального выражения
def f(x, y, z):
    return z

# Определение пределов интегрирования
x_lower = 0
x_upper = 1

y_lower = lambda x: 0
y_upper = lambda x: 1 - x

z_lower = lambda x, y: 0
z_upper = lambda x, y: 1 - x - y

# Визуализация области интегрирования
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(x_lower, x_upper, 20)
y = np.linspace(y_lower(0), y_upper(0), 20)
X, Y = np.meshgrid(x, y)
Z = 1 - X - Y

ax.plot_surface(X, Y, Z, alpha=0.5, cmap='coolwarm')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Область интегрирования')
plt.show()

# Вычисление тройного интеграла
integral, error = tplquad(f, x_lower, x_upper, y_lower, y_upper, z_lower, z_upper)

# Вывод результата
print("Значение тройного интеграла:", integral)
```

```python
import sympy as sp

# Задаем символьные переменные
x, y = sp.symbols('x y')

# Задаем функцию, которую интегрируем
f = sp.sin(x * y)

# Задаем пределы интегрирования
y1 = -x
y2 = 1 - 2 * x ** 2
x1 = 0
x2 = 1

# Вычисляем двойной интеграл
integral = sp.integrate(f, (y, y1, y2), (x, x1, x2))

# Выводим интеграл в более читаемом формате
integral_formatted = sp.latex(integral)
print(f"Интеграл: ${integral_formatted}$")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Определяем функции для поверхностей
def f1(x):
    return x ** 2

def f2(y):
    return np.sqrt(y)

def f3(x, y):
    return x * y

# Создаем фигуру и оси
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Задаем координаты x, y и z
x = np.linspace(0, 1, 100)
y = np.linspace(0, 1, 100)
X, Y = np.meshgrid(x, y)
Z1 = f1(X)
Z2 = f2(Y)
Z3 = f3(X, Y)

# Рисуем поверхности
ax.plot_surface(X, Y, Z1, cmap='magma', alpha=0.7, edgecolor='k')
ax.plot_surface(X, Y, Z2, cmap='plasma', alpha=0.7, edgecolor='k')
ax.plot_surface(X, Y, Z3, cmap='viridis', alpha=0.7, edgecolor='k')

# Добавляем тень
ax.view_init(elev=30, azim=-120)
ax.grid(False)

# Настраиваем оси
ax.set_xlim([0, 1])
ax.set_ylim([0, 1])
ax.set_zlim([0, 1])
ax.set_xlabel('X', fontsize=12)
ax.set_ylabel('Y', fontsize=12)
ax.set_zlabel('Z', fontsize=12)

# Добавляем заголовок
plt.title('3D Surface Plot', fontsize=16)

# Убираем рамку
plt.box(False)

# Показываем график
plt.show()
```

```python
import sympy as sp

# Определяем переменные
x, y, z = sp.symbols('x y z')

# Определяем функцию для интегрирования
f = x * y

# Определяем пределы интегрирования
x_limits = (0, 1)
y_limits = (0, x ** 2)
z_limits = (0, x * y)

# Вычисляем интеграл
result = sp.integrate(f, (z, *z_limits), (y, *y_limits), (x, *x_limits))

# Выводим результат
print("Результат интегрирования:")
print(result)
print()

# Добавляем оформление и дополнительную информацию
print("Оригинальный тройной интеграл:")
print("∫∫∫(x * y) dz dy dx, где")
print("x ∈ [0, 1]")
print("y ∈ [0, x^2]")
print("z ∈ [0, x * y]")
print()

print("Результат:")
print(f"Интеграл равен: {result}")
print("Это значение представляет собой объем области, ограниченной указанными пределами интегрирования.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Задаем границы области интегрирования
r = np.linspace(0, 1, 50)
o = np.linspace(0, 2 * np.pi, 50)
R, O = np.meshgrid(r, o)

X = R * np.sin(O)
Y = R * np.cos(O)
Z = (X ** 2 + Y ** 2 + X ** 2) ** 0.5  # Здесь задается функция для тройного интеграла

# Выводим область интегрирования на график в виде прозрачной поверхности
ax.plot_surface(X, Y, Z, alpha=0.3, cmap='plasma', edgecolor='none')

# Устанавливаем границы осей
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([0, 1])

# Добавляем подписи к осям и заголовок
ax.set_xlabel('X', fontsize=12)
ax.set_ylabel('Y', fontsize=12)
ax.set_zlabel('Z', fontsize=12)
plt.title('Triple Integral Region', fontsize=16)

# Добавляем цветовую панель
cbar = fig.colorbar(ax.plot_surface(X, Y, Z, alpha=0.3, cmap='plasma', edgecolor='none'))
cbar.set_label('Function Value', fontsize=12)

# Убираем рамку
ax.axis('off')

# Показываем график
plt.show()
```

```python
import sympy as sp

# Определяем символы
r, O, phi = sp.symbols('r O phi')

# Определяем координаты x, y, z
x = r * sp.sin(phi) * sp.cos(O)
y = r * sp.sin(phi) * sp.sin(O)
z = r * sp.cos(phi)

# Создаем матрицу Якобиана
J = sp.Matrix([[sp.diff(x, r), sp.diff(x, O), sp.diff(x, phi)],
               [sp.diff(y, r), sp.diff(y, O), sp.diff(y, phi)],
               [sp.diff(z, r), sp.diff(z, O), sp.diff(z, phi)]])

# Вычисляем определитель Якобиана и упрощаем его
det_J = J.det().simplify()

# Определяем функцию для интегрирования
f = (r ** 2) ** (1 / 2) * r ** 2 * sp.sin(phi)

# Вычисляем интеграл
integral = sp.integrate(f * det_J, (phi, 0, sp.pi / 2), (O, 0, 2 * sp.pi), (r, 0, 1))

# Выводим результат
print(integral)
```

<big><big><big><b> <center> ## Ответы на вопросы 

<big><big><b> 1) В среде Anaconda для вычисления двойного интеграла можно использовать функции из модуля `scipy.integrate`. Основные функции для вычисления двойного интеграла в этом модуле:

- `dblquad`: Вычисляет двойной интеграл заданной функции на прямоугольной области.
- `nquad`: Обобщенная функция для вычисления многомерных интегралов, включая двойные интегралы.

<big><big><b> 2) В среде Anaconda для вычисления тройного интеграла также можно использовать функции из модуля `scipy.integrate`. Основная функция для вычисления тройного интеграла:

- `tplquad`: Вычисляет тройной интеграл заданной функции в прямоугольной области.

<big><big><b> Эти функции из модуля `scipy.integrate` предоставляют удобные способы численного вычисления двойных и тройных интегралов в Python с использованием Anaconda.
