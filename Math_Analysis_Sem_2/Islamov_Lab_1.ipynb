```python
import numpy as np
from scipy import integrate
import sympy as sp
from IPython.display import display
```

```python
x = sp.symbols('x')
display(sp.integrate(x * sp.sin(5*x), x))
display(sp.integrate(1/((x**2 + 1)*(x - 2)**2), x))
```

```python
display(sp.integrate(sp.sqrt(1 - x**2), (x, -1, 1)))
display(sp.integrate(x * sp.exp(3*x), (x, 0, 1)))
```

```python
def find_left_int_sum(eq, a, b, n):
    summ = 0
    dxi = (b - a)/n
    for i in range(1, n + 1, 1):
        summ += eq.subs(x, (a + (i/n) - dxi)) * dxi
    print('\nИнтегральная сумма на левых концах отрезка: ',summ)
```

```python
def find_right_int_sum(eq, a, b, n):
    summ = 0
    dxi = 1/n
    for i in range(1, n + 1, 1):
        summ += eq.subs(x, (a + (i/n))) * dxi
    print('\nИнтегральная сумма на правых концах отрезка: ',summ)
```

```python
x = sp.symbols('x')
eq = x
find_right_int_sum(eq, 1, 2, 4)
find_left_int_sum(eq, 1, 2, 4)
```

```python
def lower_sum(eq, a, b, n):
    summ = 0
    dxi = 1/n
    for i in range(1, n + 1, 1):
        interv = sp.Interval((a + (i/n) - dxi), (a + (i/n)))
        summ += sp.minimum(eq, x, interv) * dxi
    print('\nНижняя сумма Дарбу: ', summ)
```

```python
def upper_sum(eq, a, b, n):
    summ = 0
    dxi = 1/n
    for i in range(1, n + 1, 1):
        interv = sp.Interval((a + (i/n) - dxi), (a + (i/n)))
        summ += sp.maximum(eq, x, interv) * dxi
    print('\nВерхняя сумма Дарбу: ', summ)
```

```python
x = sp.symbols('x')
eq = x
upper_sum(eq, 1, 2, 4)
lower_sum(eq, 1, 2, 4)
```

```python
x = sp.symbols('x')
eq = sp.exp(-x**2)
find_right_int_sum(eq, 1, 2, 1000)
find_left_int_sum(eq, 1, 2, 1000)
upper_sum(eq, 1, 2, 1000)
lower_sum(eq, 1, 2, 1000)
```

```python
def find_mid_int_sum(eq, a, b, n, l):
    summ = 0
    ln = abs(b - a)
    dxi = ln/n
    for i in range(1, n + 1, 1):
        summ += eq.subs(x, ((a + ((i/n)*ln) - dxi)) + dxi * l) * dxi
    return summ
```

```python
x = sp.symbols('x')
eq = x
print(f'Интегральная сумма при разбиении на 4 части в отношении 1/2: {find_mid_int_sum(eq, 1, 2, 4, 0.5)}')
```
