```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
```

```python
def sum_grafic(a_n, n_0):
    n_arr = np.arange(1, n_0 + 1)
    sums = [sum([a_n(n) for n in np.arange(1, n)]) for n in n_arr]
    
    fig = plt.figure(figsize=(7, 7))
    ax = fig.add_subplot()
    
    ax.plot(n_arr, sums, "b")
    
    ax.grid()
    ax.axhline(y=0, color="k")
    ax.axvline(x=0, color="k")
    
    plt.show()
```

```python
q_1 = 0.5
q_2 = 1
q_3 = 2

n_arr = np.arange(1, 15)
a_arr_1 = q_1 ** n_arr 

a_n_arr_1 = lambda n: 0.5**n
a_n_arr_2 = lambda n: 1**n
a_n_arr_3 = lambda n: 2**n

print("Члены ряда при: 1 < q < 0", end='\n')
print(*a_arr_1, sep=", ", end="\n")
print("Pяд сходится", end="\n\n")
print(f"S = {0.5/(1 - 0.5)}")
sum_grafic(a_n_arr_1, 15)


a_arr_2 = q_2 ** n_arr 
print("Члены ряда при: q = 1", end='\n')
print(*a_arr_2, sep=", ", end="\n")
print("Pяд расходится", end="\n\n")
sum_grafic(a_n_arr_2, 15)

a_arr_3 = q_3 ** n_arr 
print("Члены ряда при: q > 1", end='\n')
print(*a_arr_3, sep=", ", end="\n")
print("Pяд расходится", end="\n\n")
sum_grafic(a_n_arr_3, 15)
```

```python
q_1 = 0.5
q_2 = 1
q_3 = 2

n_arr = np.arange(1, 15)
a_arr_1 = 1 / n_arr**0.5 

a_n_arr_1 = lambda n: 1/n**0.5
a_n_arr_2 = lambda n: 1/n
a_n_arr_3 = lambda n: 1/n**2

print("Члены ряда при: 1 < q < 0", end='\n')
print(*a_arr_1, sep=", ", end="\n")
print("Pяд расходится", end="\n\n")
sum_grafic(a_n_arr_1, 15)

a_arr_2 = 1 / n_arr**1
print("Члены ряда при: q = 1", end='\n')
print(*a_arr_2, sep=", ", end="\n")
print("Pяд расходится", end="\n\n")
sum_grafic(a_n_arr_2, 15)

a_arr_3 = 1 / n_arr**2
print("Члены ряда при: q > 1", end='\n')
print(*a_arr_3, sep=", ", end="\n")
print("Pяд сходится", end="\n\n")
sum_grafic(a_n_arr_3, 15)
```

```python
def sum_grafic_and_a_n(a_n, n_0):
    n_arr = np.arange(1, n_0 + 1)
    a = [a_n(n) for n in n_arr]
    sums = [sum([a_n(n) for n in np.arange(1, n)]) for n in n_arr]
    
    fig = plt.figure(figsize=(7, 7))
    ax = fig.add_subplot()
    
    ax.plot(n_arr, sums, "r", label="sum")
    ax.plot(n_arr, a, "b", label="$a_{n}$")
    
    ax.legend()
    
    ax.grid()
    ax.axhline(y=0, color="k")
    ax.axvline(x=0, color="k")
    
    plt.show()


print("|---------------------------------------------|")
print("|     Ряд     |     Lim    |    Сходимость    |")
print("|---------------------------------------------|")
print("|  1/(n)^0.5  |      0     |    Расходится    |")
print("|---------------------------------------------|")
print("|     1/n     |      0     |    Расходится    |")
print("|---------------------------------------------|")

a1 = lambda n: 1/np.sqrt(n)
print("1/(n)^0.5", end='\n\n')
sum_grafic_and_a_n(a1, 20)

a2 = lambda n: 1/n
print("1/n", end='\n\n')
sum_grafic_and_a_n(a2, 20)
```

```python
print("A)", end="\n\n")

a1 = lambda n: 1/n + 1/(n+1)
print("1/n + 1/(n+1)", end='\n\n')
sum_grafic_and_a_n(a1, 20)

a2 = lambda n: 1/n - 1/(n+2)
print("1/n - 1/(n+2)", end='\n\n')
sum_grafic_and_a_n(a2, 20)

print("B)", end="\n\n")

a1 = lambda n: 1/n + 1/n**2
print("1/n + 1/n**2", end='\n\n')
sum_grafic_and_a_n(a1, 20)
```

```python
def sum_grafic_and_a_n_2(a_n1, a_n2, n_0):
    n_arr = np.arange(1, n_0 + 1)
    a1 = [a_n1(n) for n in n_arr]
    sums1 = [sum([a_n1(n) for n in np.arange(1, n)]) for n in n_arr]
    
    a2 = [a_n2(n) for n in n_arr]
    sums2 = [sum([a_n2(n) for n in np.arange(1, n)]) for n in n_arr]
    
    fig = plt.figure(figsize=(7, 7))
    ax = fig.add_subplot()
    
    ax.plot(n_arr, sums1, "r", label="sum1")
    ax.plot(n_arr, a1, "b", label="$a_{n1}$")
    
    ax.plot(n_arr, sums2, "g", label="sum2")
    ax.plot(n_arr, a2, "y", label="$a_{n2}$")
    
    ax.legend()
    
    ax.grid()
    ax.axhline(y=0, color="k")
    ax.axvline(x=0, color="k")
    
    plt.show()


print("1.a", end="\n\n")

n_arr = np.arange(1, 21)

print("1/(n*sqrt(n+1)*sqrt(n+2)) < 1/n**2")
print(*[f"a{i+1}: {np.round(val[0], 4)} < {np.round(val[1], 4)}" for i , val in enumerate(zip(1/(n_arr*np.sqrt(n_arr+1)*np.sqrt(n_arr+2)), 1/n_arr**2))], sep="\n")

print("\n1.b", end="\n\n")
a1, a2 = lambda n: 1/(n*sqrt(n+1)*sqrt(n+2)), lambda n: 1/n**2
sum_grafic_and_a_n_2(a1, a2, 20)

print("2.a", end="\n\n")

n_arr = np.arange(1, 21)

print("(2 + sin(n))/sqrt(n) > 1/sqrt(n)")
print(*[f"a{i+1}: {np.round(val[0], 4)} < {np.round(val[1], 4)}" for i , val in enumerate(zip((2 + np.sin(n_arr))/np.sqrt(n_arr), 1/np.sqrt(n_arr)))], sep="\n")

print("\n2.b", end="\n\n")
a1, a2 = lambda n: (2 + np.sin(n))/np.sqrt(n), lambda n: 1/np.sqrt(n)
sum_grafic_and_a_n_2(a1, a2, 20)
```

```python
def estimate_sum(sequence, epsilon):
    n = 1
    _sum = sequence(n)
    prev_sum = 0
    while abs(_sum - prev_sum) > epsilon:
        prev_sum = _sum
        n += 1
        _sum += sequence(n)
    return _sum, n
```

```python
a = lambda n: (2*n-1)/2**n

print(estimate_sum(a, 0.001)[0])
```

```python
def estimate_sum(sequence, epsilon):
    n = 1
    _sum = sequence(n)
    prev_sum = 0
    while abs(_sum - prev_sum) > epsilon:
        prev_sum = _sum
        n += 1
        _sum += sequence(n)
    return _sum, n
```

```python
a = lambda n: (-1)**(n-1)*(n/(2**n))
print(estimate_sum(a, 0.001)[0])
```
