```python
import matplotlib.pyplot as plt
import numpy as np
fig1 = plt.figure()
x = np.linspace(-1,2,500)
y1 = np.exp(2*x+1)
func1 = fig1.add_subplot(1,1,1)
plt.title('function $y = e^{2*x+1}$')
plt.xlabel('x')
plt.ylabel('y')
func1.plot(x,y1,'m.')
plt.grid()
fig1.show
```

```python
fig2 = plt.figure()
x = np.linspace(-7/3*np.pi,5/3*np.pi,500)
y2 = np.sin(x+np.pi/3)
func2 = fig2.add_subplot(1,1,1)
plt.title('function $y = sin(x+pi/3)$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
func2.plot(x,y2, 'g-')
plt.show()
```

```python
fig3 = plt.figure()
x = np.linspace(0.001,5,800)
y3 = np.log(x)
func3 = fig3.add_subplot(1,1,1)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
func3.plot(x,y3,'b-')
plt.show()
```

```python
x = np.linspace(-2*np.pi, 0, 200)
plt.figure()
plt.xlabel('x')
plt.ylabel('y')

func1, = plt.plot(x,np.sin(x),'ro-.',label='y=sin(x)')
func2, = plt.plot(x,np.sin(x) - 2,'+:',label='y=sin(x)-2')
func3, = plt.plot(x,np.sin(x) + 1,'s--',label='y=sin(x)+1')
#
plt.legend(handles = [func1, func2, func3])
plt.grid()
plt.show()
```

```python
x=np.linspace(-2.0*np.pi,2.0*np.pi,100)
handle1, = plt.plot(x,np.cos(x),label='y=cos(x)')
handle2, = plt.plot(x,np.cos(2.0*x),label='y=cos(2x)')
handle3, = plt.plot(x,np.cos(0.5*x),label='y=cos(x/2)')
plt.legend(handles=[handle1,handle2,handle3])
plt.axhline(y=0, color='k') # отобразим ось x
plt.axvline(x=0, color='k') # отобразим ось y
plt.show()
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 200)
plt.figure()
plt.xlabel('x')
plt.ylabel('y')
func1, = plt.plot(x,np.cos(x),'d-.', label='y = cos(x)')
func2, = plt.plot(x,2*np.cos(x),'*-', label='y = 2*cos(x)')
func3, = plt.plot(x,0.3*np.cos(x),'--+', label='y = 0,3*cos(x)')
func4, = plt.plot(x,-np.cos(x),':s', label='y = -cos(x)')

plt.legend(handles = [func1, func2, func3, func4])
plt.grid()
plt.show()
```
