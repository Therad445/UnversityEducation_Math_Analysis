## Упражнение 1

```python
from sympy import oo
y1 = (2*x-1)/(x-1)
y2 = (2*x-1)/abs(x-1)

sp.plot((y1,(x,0,1)),(y1,(x,1,2)), ylim = [-10,10])
print('lim = infinity')

sp.plot((y1,(x,1,2)),(y1,(x,2,3)), ylim = [1,4])
print('lim = 3')

sp.plot((y2,(x,0,1)),(y2,(x,1,2)), ylim = [-1,50])
print('lim = +infinity')
sp.plot((y2,(x,1,2)),(y2,(x,2,3)), ylim = [1,4])
print('lim = 3')
```
