```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([2+3j, 2-3j, -2+3j, -2-3j])
plt.plot()
plt.scatter(np.real(x),np.imag(x), marker="*", color="green", s=30)
plt.axis('equal')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid(True)
plt.title('Complex numbers')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.show()
```

```python
x = np.array([2+3j, 2-3j, -2+3j, -2-3j])
orig = np.zeros(4)
plt.plot()
q = plt.quiver(orig,orig, np.real(x), np.imag(x),
angles='xy', scale_units='xy', scale=1, color="green")
plt.axis([-4,4,-4,4])
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid(True)
plt.title('Complex numbers')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.show()
```

```python
z = 1 +1j
r = np.abs(z)
phi=np.angle(z)
k = np.arange(1,7)
zroot = r**(1/6)*( np.cos((phi+2.0*np.pi*k)/6) + 1j*np.sin((phi+2.0*np.pi*k)/6))


plt.title('$\sqrt[6]{1+j}$ complex roots')
plt.plot(np.real(zroot), np.imag(zroot), '*g') #    ,     ,     ,        .
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.axis('equal')
plt.grid()
t = np.linspace(0,2*np.pi,100)
plt.plot(r**(1/6)*np.cos(t),r**(1/6)*np.sin(t),':')
```

```python
#Task 5
z = 2*3**0.5 - 2j
r = np.abs(z)
n = 6**-1
phi = np.angle(z)
k = np.arange(1,5)
t = np.linspace(0,2*np.pi,100)

sqrt = r**n*(np.cos((phi+2*k*np.pi)/n) + 1j*np.sin((phi+2*k*np.pi)/n))
plt.plot(np.real(sqrt), np.imag(sqrt), '*g')
plt.plot(r**(n)*np.cos(t),r**(n)*np.sin(t),':')

plt.axis([-3,3,-3,3])
plt.axis('equal')
plt.grid()
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()



```
